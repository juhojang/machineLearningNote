#iris 데이터 이용하여 다중회귀 수행하기

```python
import sklearn.datasets
from sklearn.model_selection import train_test_split

raw_iris=sklearn.datasets.load_iris()
raw_iris
X_tr,X_te,y_tr,y_te=train_test_split(raw_iris.data,raw_iris.target,train_size=0.7)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import math
from sklearn.metrics import accuracy_score

model=LinearRegression()
model.fit(X_tr,y_tr)
print(model.coef_)
print(model.intercept_)
prediction=model.predict(X_te)
for i in range(prediction.shape[0]):
  prediction[i]=round(prediction[i],0)
  prediction[i]=int(prediction[i])
accuracy_score(prediction,y_te)
```

```python
import matplotlib.pyplot as plt
plt.scatter(prediction,y_te)
```

```python
diabet=sklearn.datasets.load_diabetes()
X=diabet.data
y=diabet.target
X_tr,X_te,y_tr,y_te=train_test_split(X,y,test_size=0.3)
model=LinearRegression()
model.fit(X_tr,y_tr)
prediction=model.predict(X_te)
y_te=np.array(y_te)
plt.ylabel("real")
plt.xlabel("prediction")
plt.xlim(50,350)
plt.ylim(50,350)
plt.scatter(prediction,y_te)
prediction=prediction.reshape(-1,1)
y_te=y_te.reshape(-1,1)
model.score(X_te,y_te)
```
